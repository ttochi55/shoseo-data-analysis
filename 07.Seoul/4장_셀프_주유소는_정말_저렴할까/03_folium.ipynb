# Folium

```python
import pandas as pd
import numpy as np

# Python Data. Leaflet.js Maps.
import folium
```

```python
df = pd.read_csv('./data/kakaomapapi.csv')
df.head(1)
```

```python
mapping = folium.Map(location=[df['lat'].mean(), df['lon'].mean()], zoom_start=13)

for i in df.index:
    folium.Marker(
        location=[df['lat'][i], df['lon'][i]],
        popup=df['이름'][i],
    ).add_to(mapping)

mapping
```
