# Folium

```python
import pandas as pd
import numpy as np

# Python Data. Leaflet.js Maps.
import folium
```

```python
df = pd.read_csv('data/openweatherapi.csv')
df.shape
```

```python
df.head()
```

```python
# 날씨에 따른 아이콘 
icons = {
    'Clear': 'map-marker',
    'Snow': 'cloud',
    'Rain': 'gift',
    'Extreme': 'info-sign',
    'Clouds': 'ok-circle',
    'Mist': 'Extreme'
}

df['wx_icon'] = ''
df['wx_icon'].loc[df['wx_main'] == 'Clear'] = icons['Clear']
df['wx_icon'].loc[df['wx_main'] == 'Clouds'] = icons['Snow']
df['wx_icon'].loc[df['wx_main'] == 'Mist'] = icons['Rain']
df['wx_icon'].loc[df['wx_main'] == 'Rain'] = icons['Extreme']
df['wx_icon'].loc[df['wx_main'] == 'Snow'] = icons['Clouds']
df['wx_icon'].loc[df['wx_main'] == 'Extreme'] = icons['Mist']
```

```python
mapping = folium.Map(location=[df['lat'].mean(), df['lon'].mean()], zoom_start=13)

# 날씨에 따른 마커의 색깔
color_dict = {
    'Clear': 'blue',
    'Snow': 'white',
    'Rain': 'gray',
    'Extreme': 'red',
    'Clouds': 'orange',
    'Mist': 'green'
}

for i in df.index:
    folium.Marker(
        location=[df['lat'][i], df['lon'][i]],
        popup=df['bdNm'][i],
        tooltip=f"{df['wx_description'][i]}, {df['wx_temp'][i]}",
        icon=folium.Icon(color=color_dict[df['wx_main'][i]], icon=df['wx_icon'][i])
    ).add_to(mapping)

mapping
```
